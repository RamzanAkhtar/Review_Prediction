```python

import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

# Load the dataset
df = pd.read_csv('TestReviews.csv')

# Split the data into training and testing sets
train_text, test_text, train_labels, test_labels = train_test_split(df['review'], df['class'], test_size=0.2, random_state=42)

# Create a tokenizer to split the text into words
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(train_text)

# Convert the text into sequences of words
train_sequences = tokenizer.texts_to_sequences(train_text)
test_sequences = tokenizer.texts_to_sequences(test_text)

# Pad the sequences to have the same length
max_length = 200
padded_train = pad_sequences(train_sequences, maxlen=max_length)
padded_test = pad_sequences(test_sequences, maxlen=max_length)

# Create the RNN model
model = Sequential()
model.add(Embedding(5000, 100, input_length=max_length))
model.add(SimpleRNN(64, dropout=0.2))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(padded_train, train_labels, epochs=10, batch_size=32, validation_data=(padded_test, test_labels))

# Evaluate the model
loss, accuracy = model.evaluate(padded_test, test_labels)
print(f'Test accuracy: {accuracy:.3f}')


#This code tests the model with two reviews: a positive review and a negative review. The output will be the predicted probabilities of the reviews being positive or negative.
```

```python

# Test the model with positive and negative reviews
positive_review = "I loved this product! It's amazing."
negative_review = "This product is terrible. I hate it."

# Preprocess the reviews
positive_sequence = tokenizer.texts_to_sequences([positive_review])
negative_sequence = tokenizer.texts_to_sequences([negative_review])

padded_positive = pad_sequences(positive_sequence, maxlen=max_length)
padded_negative = pad_sequences(negative_sequence, maxlen=max_length)

positive_prediction = model.predict(padded_positive) > 0.5
negative_prediction = model.predict(padded_negative) < 0.5

# Print the predictions
print("Positive Review:", int(positive_prediction))
print("Negative Review:", int(negative_prediction))



```

```python
random_review = "The food was okay, but the service was slow."

# Preprocess the review
random_sequence = tokenizer.texts_to_sequences([random_review])
padded_random = pad_sequences(random_sequence, maxlen=max_length)

# Get the prediction
random_prediction = model.predict(padded_random) > 0.5

# Print the prediction
print("Random Review:", int(random_prediction))

```
